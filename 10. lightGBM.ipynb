```python
def replace_months(day):
    if day=='nan':
        print('nan')
        return 'nan'
    print(day)
    datas = day.split(' ')
    result = int(datas[0]) * 12 + int(datas[-2])

    return result
```

```python
# 모든 대출이름을 파생변수로 생성
# 해당 대출이 발생한 경우 1 아니면 0으로 채움
```

```python
# and 없애기
credit_df['Type_of_Loan'] = credit_df['Type_of_Loan'].str.replace('and ', '')
```

```python
credit_df.head()
```

```python
# ', '없애기
credit_df['Type_of_Loan'] = credit_df['Type_of_Loan'].str.replace(', ', ',')
# ',  '없애기
credit_df['Type_of_Loan'] = credit_df['Type_of_Loan'].str.replace(',  ', ',')
```

```python
credit_df.head()
```

```python
# nan일 경우 문자열 none 처리
credit_df['Type_of_Loan'] = credit_df['Type_of_Loan'].fillna('none')
```

```python
# 대출 유형 리스트
type_list = []
```

```python
# 반복문 돌리면서 리스트에 담기
for i in credit_df['Type_of_Loan']:
    #
    if type(i) == float:
        continue

    # , 기준으로 나눔
    datas = i.split(',')
    # type_list에 요소들 추가
    type_list.extend(datas)
```

```python
# set으로 변경 후 중복값 제거
type_set = set(type_list)
```

```python
# set 출력
type_set
```

```python
# 다시 리스트로 변경
type_list = list(type_set)
print(type_list)
```

```python
# 반복문 돌리면서 대출 유형 파생변수 추가
# apply 함수를 이용하여 문자열을 포함하고 있으면 1, 없으면 0으로 채우기
for i in type_list:
    credit_df[i] = credit_df["Type_of_Loan"].apply(lambda x: 1 if i in x else 0)
```
