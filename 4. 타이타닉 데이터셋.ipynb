# 2. 데이터 전처리
* 넓은 범위의 데이터 정제 작업을 뜻함
* 필요없는 데이터를 삭제하고, 필요한 데이터만 취하는 것, null 값이 있는 행을 삭제하는 것, 정규화, 표준화 등의 많은 작업들을 포함하고 있음
* 머신러닝, 딥러닝 실무에서도 전처리가 50% 이상의 중요도를 차지함

### 2-2. 결측치 처리

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# 나이 결측치를 평균으로 처리
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age']
```

### 2-3. 라벨 인코딩(Label Encoding)
* 문자(Categorycal)를 수치(Numerical)로 변환

```python
df.info()
```

```python
df['Sex'].value_counts()
```

```python
def conver_sex(data):
    if data =='male':
        return 1
    elif data == 'female':
        return 0
```

```python
df['Sex'] = df['Sex'].apply(conver_sex)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Embarked'].value_counts()
```

```python
le.fit_transform(df['Embarked'])
```

```python
le.classes_ # 0: 'C', 1: 'Q', 2: 'S', 3: nan
```

### 2-4. 원 핫 인코딩(One Hot Encoding)
* 독립적인 데이터는 별도의 컬럼으로 분리하고 각각 컬럼에 해당 값에만 1, 나머지는 0값을 갖게 하는 방법
* 예) 머신러닝 알고리즘은 'c:0', 'Q:1', 'S:2', 'NaN:3' 데이터의 관계성을 찾아 'Q+Q=S'라고 학습을 할 수 있음 -> 관계성을 끊어주기 위해 원 핫 인코딩을 사용

```python
df['Embarked_num'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
pd.get_dummies(df['Embarked_num'])
```
